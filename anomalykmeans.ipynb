```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
import scipy.io
import seaborn as sns
```

# Loading DataSet

This module import the required dataset and comvert into numpy format

```python
Small_data_set = scipy.io.loadmat('ex8data1.mat')
Large_data_set =  scipy.io.loadmat('ex8data2.mat')
small_data = Small_data_set['X']
cross_validation_xval=Small_data_set['Xval']
yval=Small_data_set['yval']
small_data = np.array(small_data)
xval= np.array(cross_validation_xval)
yval=np.array(yval)
large_data = Large_data_set['X']
large_data = np.array(large_data)
print(small_data)
```

## Plotting the data

Code below plots the small dataset using in 2-dimensional plane using matplotlib library

```python
df = pd.DataFrame(data=small_data)
sns.set_style('darkgrid')
sns.FacetGrid(df,size=8).map(plt.scatter,0,1,label='2d scatter plot').add_legend()
plt.xlabel('throughput(mb/s)')
plt.ylabel('Latency')
plt.show()
#df
```

## Shape of dataset

```python
print(small_data.shape)
print(large_data.shape)
```

```python
plt.close()
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=5000)
kmeans.fit(small_data)
y = kmeans.predict(small_data)
plt.scatter(small_data[:, 0], small_data[:, 1], c=y, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
#plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.show()
```
